```python
#data_3.loc[:, "Country/Region"]
#Dates of the confirmed cases, recoveries and deaths of COVID-19 in sync with the countries of choice
def case_date():
    date_list = []
    #formating the dates from the country of choice to have
    for index in countries_of_choice.iloc[:, 41::4]:
        #spliting the dates into lists of ['month', 'day', 'year']
        month, day, year = index.split('/')
        f_date1 = month, day, year + '20'
        f_date2 = '/'.join(f_date1)
        date_list.append(f_date2)
    print(date_list)
case_date()

#print(dir(index))
```

### Line Graph

```python
#Use this as well as the next line of code to increase plot canvas size (plt.rcParams['figure.figsize'] = [30, 20])
plt.figure(figsize=(20, 15))
plt.plot(date_time, cases_Australia, color = 'yellow', label = "Australia", linewidth=3)
plt.plot(date_time, cases_Benin, color = 'red', label = "Benin", linewidth=3)
plt.plot(date_time, cases_Bhutan, color = 'blue', label = "Bhutan", linewidth=3)

plt.title("Confirmed Cases of COVID-19")
plt.legend(shadow="True")
plt.show()
```

```python
plt.figure(figsize=(20, 15))
plt.plot(date_time, deaths_Australia, color = 'yellow', label = "Australia", linewidth=3)
plt.plot(date_time, deaths_Benin, color = 'red', label = "Benin", linewidth=3)
plt.plot(date_time, deaths_Bhutan, color = 'blue', label = "Bhutan", linewidth=3)

plt.title("Death Cases of COVID-19")
plt.legend(shadow="True")
plt.show()
```

### Grouped Bar Chart

```python
plt.figure(figsize=(10, 5))
N = 3
index = np.arange(N)
width = 0.3
label_1 = "cases", "recoveries", "deaths" 

total_Australia = [cases_Australia[-1], recoveries_Australia[-1], deaths_Australia[-1]]
total_Benin = [cases_Benin[-1], recoveries_Benin[-1], deaths_Benin[-1]]
total_Bhutan = [cases_Bhutan[-1], recoveries_Bhutan[-1], deaths_Bhutan[-1]]


plt.bar(index - 0.25, total_Australia, color = 'yellow',  width=width, label = "Australia")
plt.bar(index - 0.25 * -0.26, total_Benin, color = 'red',  width=width, label = "Benin")
plt.bar(index + 0.3, total_Bhutan, color = 'blue',  width=width, label = "Bhutan")

plt.xticks(index - 0.1, label_1)
plt.title("Confirmed Cases of COVID-19")
plt.legend(loc='upper right', shadow="True")
plt.show()
```
