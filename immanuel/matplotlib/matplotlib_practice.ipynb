```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
plt.style.use("ggplot")
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50,51,52,48,47,49,46]
z = [45,47,52,55,50,51,60]
```

```python
#"r^-" red triangles joined with lines
plt.plot(x, y, "r^-")
plt.scatter(x, y)
plt.show()
```

```python
data = pd.DataFrame({"x":x, "y":y})
sns.lmplot("x", "y", data=data, fit_reg=True)
```

```python
plt.xlabel("Night")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(x, y, color="green", linewidth=4, alpha=1, label="Owerri")
plt.plot(x, z, color="pink", linewidth=2, alpha=1, label="Eko")
plt.legend()
plt.show()
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.style.use("ggplot")
#"bs--" blue square dashes,for more refer to https://matplotlib.org/tutorials/introductory/pyplot.html
#plt.subplot(131), plt.subplot(132)... aids categorical plotting, plotting values side by side
plt.subplot(131)
plt.plot(x, z, "bs--")
plt.subplot(132)
plt.bar(x, z, color="red")

plt.show()
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y)
plt.plot(x-0.2, z)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, color="blue", label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="avg")
plt.legend()
```

#### CREATING GRID STYLE MULTIPLE PLOTS IN SINGLE FIGURE

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 5))
fig.suptitle("TEST FOR STUDENTS")
plot1 = axs[0, 1]
plot2 = axs[1, 1]
plot3 = axs[1, 0]
```

```python
fig, axs = plt.subplots(1, 4, figsize=(20,4))

plot1 = axs [0]
plot2 = axs [1]
plot3 = axs [2]
plot4 = axs [3]

plot1.set_title("Math Scores")
plot2.set_title("English Scores")
plot3.set_title("French Scores")
plot4.set_title("Combined Scores")

study_hrs =[1,2,3,4,5,6,7]
math=[50,51,52,48,47,49,46]
english=[43,42,40,44,33,35,37]
french=[45,48,48,46,40,42,41]

plot1.plot(study_hrs, math, color="green")
plot2.plot(study_hrs, english, color="orange")
plot3.plot(study_hrs, french, color="blue")

plot4.plot(study_hrs, math, color="green", label="Math")
plot4.plot(study_hrs, english, color="orange", label="English")
plot4.plot(study_hrs, french, color="blue", label="French")

plot4.legend()
plt.show()
```

```python
number_of_values = 5
menMeans = np.array((20, 35, 30, 35, 27))
men = ["Ade", "shayo", "Bolu", "Gide", "Jemile"]
subject = ["Math", "Science", "History", "Chemistry", "Biology"]
womenMeans = np.array((25, 32, 34, 20, 25))
womenMeans2 = np.array((15, 39, 30, 20, 19))
index = np.arange(number_of_values)  #the index location of the 5 values that could be seen in "men"
width = 0.35
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
#Sbtracting from the index to accomodate another plotting along side it
plt.bar(index - 0.25, menMeans, color=("y","r", "g"), label="Mean Scores", width=width)
#Adding to the index to align the bars properly alongside each other
plt.bar(index + 0.25, menStd, color=("c", "r", "g"), label="Standard Deviation", width=width)
plt.plot(index - 0.25, menStd, color="black", label="Score Std")

#use the plt.xticks to display the names instead of the index
plt.xticks(index - 0.15, men)

plt.legend(loc = "upper right")
plt.title("Student Scores (mean and standard deviation)")
plt.xlabel("Student Name")
plt.ylabel("Student Scores")
plt.show()
```
