```python
import pandas as pd
import matplotlib
```

```python
#Creating tables in panda using a dictionary
data = dict(names = ["Foo", "Bar", "Lorem", "Ipsum", "Dade"],
           age = [20, 30, 40, 35, 50],
           hobbies = ["dancing", "playing scrabble", "running", "reading", "sleeping"])
new_df = pd.DataFrame(data)
data
```

```python
new_df
```

```python
#Creating a table in pandas by assigning values ("df"==dataframe)
new_df1 = pd.DataFrame()
```

```python
new_df1["names"] = ['Foo', 'Bar', 'Lorem', 'Ipsum', 'Dade']
new_df1["gender"] = ["male", "female", "male", "female", "male"]
new_df1["hobbies"] = ["dancing", "playing scrabble", "running", "reading", "sleeping"]
new_df1
```

```python
#Reading a csv file with pandas
pd.read_csv("time_series_covid19_deaths_global.csv")
```

```python
#To obtain technical information/technical overview about a DataFrame by using the ".info()" method 
new_df.info()
```

```python
#To get a statistical overview of a DataFrame we use the ".describe()" method
new_df.describe()
```

```python
#Drop a colum using ".drop(columns = ["name of column"])"
new_df1.drop(columns = ["gender"])
```

```python
#To find the max age in the dataset
#using the ".max()" method without passing in age will pull out all the details of the person with the maximum age
new_df.age.max()
```

```python
#To find the min age in the dataset
#using the ".min()" method without passing in age will pull out all the details of the person with the maximum age
new_df.age.min()
```

```python
#booleans in pandas
new_df.age == 50
```

```python
#This is used to filter the boolean return value, to get just the person's name, add ".names" after the bracket or whatever 
#you named the column housing the list of names
new_df[new_df.age==50].names
```

```python
#Looking for the max using a more implicit approach
max_age = new_df.age.max()
new_df[new_df.age==max_age]
```

```python
#Plotting your data on a chart with pandas using matplotlib
new_df.plot("names", "age", kind="bar", color=["red", "blue", "green", "purple"])
```
